***Теория Вероятности*** - теоретическая оценка случайного события, а *Статистика* - работает с практическими/империческими данными.

Методы статистической оценки:
1. Оценка неизвестной вероятности, функции распределения либо
   оценка параметров распределения, вид которого неизвестен;
2. Проверка статистических гипотез о виде неизвестного распределения
   или о величине параметров распределения, чей вид известен.

***Цель математической статистики*** - оценка характеристик генеральной совокупности (ГС) по выборочным данным.

Свойства ГС обычно неизвестны и все выводы о ГС делаются исключительно по выборке.

***ГС***- совокупность всех возможных объектов данного вида, над которыми проводятся
наблюдения с целью получения конкретных значений опред случ величины.

***Выборка/выб совокупность*** - совокупность случайно отобранных объектов из ГС.

Описательная статистика (ОС) отвечает на вопрос "как лучшим образом сжато описать данные?"
В  ОС выделяют меры центральной тенденции (меры концентрации):
- медиана;
- мода;
- среднее арифметическое (СА);
- перцентили.

***Медиана*** - такое число выборки, что ровно половина из элементов выборки больше него, а другая - меньше.
Медиана - серединное значение, данные упорядочены по убыванию или возрастанию.

Медиана лучше СА в качестве меры центральной тенденции для симметричных данных или данных с выбросом.
Это связано с тем, что медиана основана на рангах, а не самих значениях.
И по определению половина значений лежит ниже медианного значения, а половина - выше
независимо от конкретных чисел.

***Квантиль*** - значение, которое заданная случайная величина не превышает с фиксированной вероятностью.
Если вероятность задана в процентах - квантиль называется процентилью/перцентилью.

Разброс говорит о том, на сколько значения рассеяны в данных.
По аналогии с мерами концентрации в зависимости от того,
по ГС или по выборке производится оценка, влияют разные меры разброса.
Для ГС данных используют дисперсию (квадрат среднеквадр отклонения (СО)).

По интуиции, разброс значений внутри выборки не такой большой, как в ГС
и деление не на n, а на n-1 позволяет скомпенсировать это различие.

При оценке ГС через выборку, мы можем ошибаться - такая ошибка называется стандартная,
она расчитывается для любого показателя, и чаще всего используется стандартная ошибка среднего (СОС,
стандартное отклонение СА), то есть она оценивает точность, с которой выборочная средняя оценивает среднее ГС.

***СОС*** - величина, харак-ая стандартное отклонение выборочного среднего,
расчитанное по выборке размера n из генеральной совокупности (s / sqrt(n))

Чем больше выборка, тем точнее оценка среднего и тем меньше его стандартная ошибка.

В отличие от ***стандартного отклонения стандартная ошибка среднего*** ничего не говорит о разбросе данных,
оценкона лишь показывает точность выборочной оценки среднего.

Различают:
- точечные оценки - значение неизвестного параметра приближается отдельным числом;
- доверительные интервалы - случайный интерал, построенный по выборке, которы содержит истинное значение параметра с заданной веротностью.

Основные свойства точечных оценок:
- несмещенность - выб показатель, который при многократном повторении выборки стремится к теор значению.
  Потому что мы не знаем реальных значений показателей по ГС и с помощью выб наблюдений пытаемся его оценить.
  Оценка показателя и есть его характеристика, расчитанная по выборке. При этом выб среднее - несмещенная ооценка мат ожидания.
  Так как средняя из выб средних стремтся к своему теор значению по ГС.
  
- состоятельность - при бесконечном расширении выборки оценка приходит к истинному значению.

- эффективность. Несмещенная оценка наз ***эффективной*** среди рассм оценок, если у нее минимальная дисперсия.


Оценка называется ***несмещенной***, если ее мат ожидание равно истинному значению оцениваемого параметра.
Если мат ожидание оценки сходитсяк истинному значению параметра с ростом объема выборки - ***асимптот немиещенность.***

Для работы с оценкой статистических критериев используется ряд функций:
- **pdf** - функция плотности вероятности - показывает вероятность распределения случайной величины;
- **cdf** - кумулятивная функция распределения дает интегральную картину распределения вероятности,
  задает вопрос: "Какова вероятность того, что результат окажется меньше или равен такому-то?"

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats


np.random.seed(1234)
```

***<center>Distribution</center>***

```python
# формирование ГС с норм распр
norm_rvl = stats.norm(loc=35, scale=10)
# scale - standard deviation
# loc - average

# generate random values from norm_rvl
gen_pop = norm_rvl.rvs(size=10000)

fig1 = plt.figure(figsize=(10, 5))
ax1 = plt.subplot(111)
plt.hist(gen_pop, 50, alpha=.5)
plt.show()

# формирование ГС с логнорм распр
ln_distr = stats.lognorm(.5, loc=25, scale=10)
gen_pop_ln = ln_distr.rvs(size=10000)

fig2 = plt.figure(figsize=(10, 5))
ax2 = plt.subplot(111)
plt.hist(gen_pop_ln, 100, alpha=.5)
plt.show()
```

***<center>Median</center>***

```python
median = np.median(gen_pop)
median_ln = np.median(gen_pop_ln)

fig = plt.figure(figsize=(10, 5))
ax1 = plt.subplot(111)
plt.hist(gen_pop, 50, alpha=.8)
linel = plt.axvline(median, label=(f'Median = {round(median, 1)}'), color='m', linestyle='dashed', linewidth=3.5)

ax1.legend(handles=[linel], fontsize=18)
plt.show()

fig = plt.figure(figsize=(10, 5))
ax1 = plt.subplot(111)
plt.hist(gen_pop_ln, 50, alpha=.8)
linel = plt.axvline(median_ln, label=(f'Median = {round(median_ln, 1)}'), color='m', linestyle='dashed', linewidth=3.5)

ax1.legend(handles=[linel], fontsize=18)
plt.show()
```

***<center>Mode</center>***

```python
dscrt_list = np.random.randint(0, 100, size=1000)

fig = plt.figure(figsize=(10, 5))
ax1 = plt.subplot(111)
plt.hist(dscrt_list, 100, alpha=.8)
plt.show()

mode = stats.mode(dscrt_list)
print(mode)
```

***<center>Percentile</center>***

```python
res = np.percentile(gen_pop, 75)
res = stats.scoreatpercentile(gen_pop, 75)
print(res)

df_box = pd.DataFrame()
df_box['gen_pop'] = gen_pop
df_box['gen_pop_ln'] = gen_pop_ln

fig = plt.figure(figsize=(10, 5))
sns.boxplot(data=df_box, orient='h')
plt.show()
```

***<center>Average/Mean</center>***

```python
mean = np.mean(gen_pop)

n = 20
sample = np.random.choice(gen_pop, n, replace=False)
sample_mean = np.mean(sample)
print(sample_mean)
```

***<center>Scatter Measures</center>***

```python
var = np.var(gen_pop)
std_ = np.std(gen_pop)
print(var, std_)

# оценка СО через выборочное СО
std_sample = np.std(sample, ddof=1)
print(std_sample)

iteration = 1000
n = 20

std_sample_l = []
std_sample_l_corrected = []

for i in range(iteration):
    sample = np.random.choice(gen_pop, n, replace=False)
    std_sample = np.std(sample)
    std_sample_corrected = np.std(sample, ddof=1)

    std_sample_l.append(std_sample)
    std_sample_l_corrected.append(std_sample_corrected)

fid = plt.figure(figsize=(10, 5))
ax1 = plt.subplot(111)
plt.hist(std_sample_l, color='cyan', alpha=.5, bins=30)
plt.hist(std_sample_l_corrected, color='magenta', alpha=.5, bins=30)
line1 = plt.axvline(std_, color='black', linestyle='dashed', linewidth=4.5, label='std ГС')
line2 = plt.axvline(np.mean(std_sample_l), color='blue', linestyle='dashed', linewidth=4.5, label='std выборки')
line3 = plt.axvline(np.mean(std_sample_l_corrected), color='red', linestyle='dashed', linewidth=4.5, label='std выборки скорректированное')

plt.legend()
plt.show()
```

***<center>Standard Error of the Average</center>***

```python
n = 20
sample = np.random.choice(gen_pop, n, replace=False)
# оценка СОС по выборке
print(stats.sem(sample))

fig = plt.figure(figsize=(10, 5))

ax1 = plt.subplot(111)
plt.hist(gen_pop, 50, alpha=.8)
line1 = plt.axvline(mean, label=f'Mean = {mean:.1f}', color='magenta', linestyle='dashed', linewidth=3.5, alpha=.8)
ax1.legend(handles=[line1], fontsize=15)

plt.show()
```

```python
n = 100
for i in range(5):
    sample = np.random.choice(gen_pop, n, replace=False)
    sample_mean =np.mean(sample)
    sem = stats.sem(sample)
    fig = plt.figure(figsize=(10, 5))
    
    ax1 = plt.subplot(111)
    plt.hist(sample, 20, alpha=.8)
    line1 = plt.axvline(sample_mean, label=f'Выборочное среднее = {sample_mean:.1f}',
                        color='blue', linestyle='dashed', linewidth=3, alpha=.4)
    line2 = plt.axvline(mean, label=f'Истинное среднее = {mean:.1f}',
                        color='magenta', linestyle='dashed', linewidth=3, alpha=.4)
    line3 = plt.axvline(mean, label=f'Выборочная стандартная ошибка среднего = {sem:.1f}',
                        color='black', linestyle='dashed', linewidth=3, alpha=0)

ax1.legend(handles=[line1, line2, line3], fontsize=10)
plt.show()
```

```python
iteration = 5000
n = 20

std_sample_l = []
std_sample_l_corrected = []

est = []
est_l = []

for i in range(iteration):
    sample = np.random.choice(gen_pop, n, replace=False)
    std_sample = np.std(sample)
    std_sample_corrected = np.std(sample, ddof=1)

    std_sample_l.append(std_sample)
    std_sample_l_corrected.append(std_sample_corrected)
    
    if i % 20 == 0:
        est.append(np.abs(np.mean(std_sample_l) - std_))
        est_l.append(np.abs(np.mean(std_sample_l_corrected) - std_))
```

```python
fig = plt.figure(figsize=(12, 6))
plt.plot(est)
plt.plot(est_l)
plt.show()
```

```python
est = []
for i in range(10, 5000, 10):
    sample = np.mean(np.random.choice(gen_pop, i, replace=False))
    est.append(sample)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(est, c='grey', alpha=.8)
plt.hlines(mean, 0, 500, color='magenta', lw=2, label='Среднее ГС')
plt.xlabel('Кол-во наблюдений * 10', size=12)
plt.legend()
plt.show()
```

```python
# Пример несостоятельной оценки
est = []
for i in range(10, 5000, 10):
    sample = np.mean(np.random.choice(gen_pop, i, replace=False)) * (np.sqrt(i)/i) + 30
    est.append(sample)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(est, c='grey', alpha=.8)
plt.hlines(mean, 0, 500, color='magenta', lw=2, label='Среднее ГС')
plt.xlabel('Кол-во наблюдений * 10', size=12)
plt.legend()
plt.show()
```

```python
n = 20
est1 = []
est2 = []

for i in range(1000):
    est1.append(np.mean(np.random.choice(gen_pop, n, replace=False)))
    est2.append(np.median(np.random.choice(gen_pop, n, replace=False)))

df_box_plot_est = pd.DataFrame()
df_box_plot_est['est1'] = est1
df_box_plot_est['est2'] = est2
```

```python
fig = plt.figure(figsize=(12, 6))
sns.boxplot(data=df_box_plot_est, orient='h', showmeans=True)
plt.show()
```

```python
mean1 = np.mean(est1)
mean2 = np.mean(est2)

print(f'Среднее по выборкам оценки 1 - {mean1:.1f}')
print(f'Среднее по выборкам оценки 2 - {mean2:.1f}')

var1 = np.var(est1)
var2 = np.var(est2)

print(f'Дисперсия по выборкам оценки 1 - {var1:.1f}')
print(f'Дисперсия по выборкам оценки 2 - {var2:.1f}')
```

Обе оценки состояельны и несмещенные, при этом оценка среднего ГС через СА ***эффективнее***, так как ее дисперсия ***меньше***.

```python
fig = plt.figure(figsize=(12, 6))
x = np.linspace(gen_pop.min(), gen_pop.max(), 100)
mu = gen_pop.mean()
sigma = gen_pop.std()
y_pdf = stats.norm.pdf(x, mu, sigma)
y_cdf = stats.norm.cdf(x, mu, sigma)

plt.plot(x, y_cdf, label='cdf', color='r')
plt.legend()
plt.twinx()
plt.plot(x, y_pdf, label='pdf', color='b')
plt.legend()
plt.show()
```
