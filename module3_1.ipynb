```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig = plt.figure(figsize=(8, 4))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(122)
ax3 = fig.add_subplot(223)
# ax4 = fig.add_subplot(224)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 4))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(122)
ax3 = fig.add_subplot(223)
ax1.hist(np.random.randn(100), bins=30, alpha=1)
ax2.scatter(np.arange(30), np.arange(30) + 5 * np.random.randn(30))
ax3.plot(np.random.randn(30).cumsum())

# plt.show()
```

```python
fig, axes = plt.subplots(2, 3)
fig.set_figheight(5)
fig.set_figwidth(10)

axes[0, 1].scatter(np.arange(30), np.arange(30) - 2 * np.random.randn(30))
axes[1, 1].bar(np.arange(10), np.random.rand(10))

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(10)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='m', alpha=1)

# Create space around graphs
plt.subplots_adjust(wspace=0.0, hspace=0.0)

plt.show()
```
