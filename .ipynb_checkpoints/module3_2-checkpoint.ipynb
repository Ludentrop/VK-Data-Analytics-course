```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# pd.set_option('display.max_columns', None)

df = pd.read_csv('data/heart.csv')
```

```python
print(df.head())
```

```python
fig = plt.figure(figsize=(10, 5))
df['ChestPainType'].value_counts().plot.bar()
df['ChestPainType'].value_counts().sort_index().plot.bar()
(df['ChestPainType'].value_counts() / len(df)).plot.bar()
res = df.groupby('HeartDisease')[['Age', 'Cholesterol']].mean()  # = df.groupby('HeartDisease').agg({'Age': 'mean', 'Cholesterol': 'mean'})
print(res)
```

```python
df.groupby('HeartDisease')[['Age', 'Cholesterol']].mean().plot.bar(figsize=(10, 5), stacked=False)
```

```python
df['Cholesterol_bin'] = df['Cholesterol'] // 10 * 10
fig = plt.figure(figsize=(10, 5))
chart = df['Cholesterol_bin'].value_counts().sort_index().plot.line()
df['Cholesterol_bin'].value_counts().sort_index().plot.area(alpha=.5)
```

```python
df['Age'].plot.hist(bins=10)
```

```python
df.plot.scatter(x='Cholesterol', y='Age', figsize=(10, 5))
```

```python
df['Cholesterol'] = np.where(df['Cholesterol']==0, np.nan, df['Cholesterol'])
df.plot.hexbin(x='Cholesterol', y='Age', figsize=(10, 5), gridsize=20, cmap='Blues')
```
